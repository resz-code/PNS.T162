Cell 1: Memanggil Library (Wajib dijalankan pertama kali)

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import joblib

print("Library berhasil dimuat!")
```

Cell 2: Menyiapkan Data & Preprocessing

```python
# 1. Menyiapkan data historis (Baseline)
# Fitur: [Suhu_Mesin, Getaran_Mesin]
X_train = np.array([[60, 2], [70, 4], [80, 6], [90, 8], [100, 10]])
y_train = np.array([10, 25, 45, 70, 95]) # Skor Risiko

# 2. Preprocessing (Scaler)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

print("Data berhasil disiapkan dan di-scale!")
```

Cell 3: Melatih dan Menyimpan Model (Persistensi)

```python
# 3. Training Model
model = LinearRegression().fit(X_train_scaled, y_train)
print("Model berhasil dilatih.")

# 4. Menyimpan model dan scaler ke dalam file .joblib
joblib.dump(model, 'model_risiko_v1.joblib')
joblib.dump(scaler, 'scaler_risiko_v1.joblib')

print("SUKSES: File 'model_risiko_v1.joblib' dan 'scaler_risiko_v1.joblib' telah tersimpan di folder kerjamu.")
```

Cell 4: Fungsi Tambahan (Syarat Praktikum)

```python
# Fungsi Deteksi Drift (Memeriksa apakah data baru menyimpang)
def check_data_drift(new_data, train_mean, threshold=2.0):
    drift = np.abs(np.mean(new_data) - np.mean(train_mean))
    if drift > threshold:
        return f"WARNING: Terdeteksi Drift sebesar {drift:.2f}. Segera kalibrasi model!"
    return "Sistem Stabil: Data input sesuai dengan profil model."

# Fungsi Anonimisasi Data (Menghapus data pribadi/sensitif)
def clean_sensitive_data(df_input):
    cols_to_remove = ['Nama_Operator', 'NIK_Petugas']
    # Hanya hapus kolom jika kolom tersebut ada di dalam dataframe
    df_cleaned = df_input.drop(columns=[c for c in cols_to_remove if c in df_input.columns])
    return df_cleaned

print("Fungsi Deteksi Drift dan Anonimisasi siap digunakan.")
```
